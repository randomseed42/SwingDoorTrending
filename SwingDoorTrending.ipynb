```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from SwingDoorTrending import swing_door_trending
```

```python
np.random.seed(0)
N = 100
step = 1
x = np.arange(0, N*step, step)
y = np.random.randint(0, 10, size=(N))

_, x_new, y_new = swing_door_trending(x, y, 0.01, m=(15,85))

fig, ax = plt.subplots(figsize=(15,3))
ax.scatter(x, y)
ax.plot(x, y)
ax.scatter(x_new, y_new)
ax.plot(x_new, y_new)
plt.tight_layout()
plt.show()
```

```python
N = 1000
step = 1
x = np.arange(0, N*step, step)
y = np.sin(x*np.pi/180)

_, x_new, y_new = swing_door_trending(x, y, 0.001, m=(15,85))

fig, ax = plt.subplots(nrows=2, figsize=(15,6))
ax[0].scatter(x, y)
ax[0].plot(x, y)
ax[1].scatter(x_new, y_new)
ax[1].plot(x_new, y_new)
plt.tight_layout()
plt.show()
```
